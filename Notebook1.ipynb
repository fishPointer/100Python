```python
# Day 5 Project - Random Password Generator
import random

typelist = ["C", "S", "N"]
charlist = [chr(97+x) for x in range(26)] + [chr(65+z) for z in range(26)]
symlist = [chr(33+y) for y in range(15)]
numlist = [chr(48+w) for w in range(10)]
password = []

print("Welcome to password generator")
charflag = int(input("How many letters in the password?\n"))
numflag = int(input("How many numbers in the password?\n"))
symflag = int(input("How many symbols in the password?\n"))

passlength = charflag + numflag + symflag
n=0

while n < passlength:
    pick = random.choice(typelist)
    if pick == "C":
        if charflag > 0:
            password.append(random.choice(charlist))
            charflag -= 1
        else:
            n -= 1
    if pick == "S":
        if symflag > 0:
            password.append(random.choice(symlist))
            symflag -= 1
        else:
            n -= 1
    if pick == "N":
        if numflag > 0:
            password.append(random.choice(numlist))
            numflag -= 1
        else:
            n -= 1
    n += 1

password = "".join(password)
print(password)

# This can be further optimized. 
# I can use password += the random choice, so it's an array the whole time and I don't need to convert from list
# I mean technically the whole thing can be tremendously shorter if I just .shuffle() it (requires list)
# I'm sure there's a way to delete an entry from the typelist once the flag is reached, so there's no wasted loops
# It's probably worthwhile to rewrite this further down. Revisit the code afterwards during the review/summary and rewrite earlier projects
```

```python
# Step 2
# 1 - Create an empty list with one "_" for each letter
# 2 - Loop through each position in the word, and if it matches, reveal the letter in the _ list
# 3 - Print the _ list to render the game state

import random

word_list = [
    "Bicycle", "Kangaroo", "Elephant", "Butterfly", "Computer",
    "Sunshine", "Chocolate", "Mountain", "Umbrella", "Earring",
    "Telephone", "Rainbow", "Backpack", "Octopus", "Fireworks",
    "Guitar", "Calendar", "Icecream", "Lighthouse", "Watermelon",
    "Zucchini", "Keyboard", "Telescope", "Pineapple", "Dragonfly",
    "Sunflower", "Universe", "Elephant", "Starfish", "Hamburger",
    "Rhinoceros", "Snowflake", "Helicopter", "Giraffe", "Basketball",
    "Lighthouse", "Strawberry", "Camouflage", "Waterfall", "Kangaroo",
    "Peacock", "Scarecrow", "Rainbow", "Dragonfly", "Pineapple",
    "Jellyfish", "Campfire", "Snowboard", "Football", "Telephone"
]


print("Hangman Simulator Part 2")
user = input("Guess a letter: ")
user = user.upper()
print(user)


gameword = random.choice(word_list)
gameword = gameword.upper()
print(gameword)

gamelength = len(gameword)
print(f"Word Length: {gamelength}")

gamestate = []

# Note: you cannot pass an int into the for loop. It has to be range(int) to generate the list

for x in range(gamelength):
    gamestate.append("_")

print(f"Game State: {gamestate}")

print("Checking for letter matches.")
print(f"User input: {user}")
for x in range(gamelength):
    print(f"Letter #{x}: {gameword[x]} vs {user}")
    if gameword[x] == user:
        print("True")
        gamestate[x] = gameword[x]
    else:
        print("False")

print(f"New Game State: {gamestate}")
```

```python
# Generated by chatgpt.
# Increment upwards from 2
# define a range stepping from [N, Limit] moving in steps of N
# And mark all those numbers as non-prime, since they are composite.
# Increment N and repeat

def sieve_of_eratosthenes(limit):
    # Create a boolean array "is_prime" and initialize
    # all entries as True. A value in is_prime[i]
    # will finally be False if i is not a prime, else True.
    is_prime = [True] * (limit + 1)
    is_prime[0] = is_prime[1] = False  # 0 and 1 are not prime numbers

    # Start with the first prime number, 2
    current_prime = 2

    while current_prime * current_prime <= limit:
        # If is_prime[current_prime] is still True, it's a prime
        if is_prime[current_prime]:
            # Mark all multiples of current_prime as not prime
            for multiple in range(current_prime * current_prime, limit + 1, current_prime):
                is_prime[multiple] = False
        # Move to the next number
        current_prime += 1

    # Return a list of prime numbers
    primes = [i for i in range(2, limit + 1) if is_prime[i]]
    return primes

# Example usage
limit = 10000
primes = sieve_of_eratosthenes(limit)
print(primes)
```

```python
# Day 8 Final Project - Caesar Cipher
 
alphabet = [chr(ord("A") + i) for i in range(26)]
cipher = []

def shiftfix(index):
    while index > 25:
        index -= 26
    while index < 0:
        index += 26
    return index

def caesar(plaintext, shiftvalue, direction):
    if direction == "D":
        shiftvalue *= -1

    for x in range(len(plaintext)):
        index1 = alphabet.index(plaintext[x])
        index1 += shiftvalue
        index2 = shiftfix(index1)
        cipher.append(alphabet[index2])

    final = "".join(cipher)
    return final

def program():
    direction = input("Type 'E' to Encode and 'D' to Decode\n")
    text = input("Type your message:\n").upper()
    shift = int(input("Type the shift number:\n"))
    output = caesar(text, shift, direction)
    print(output)
    loop = input("Go again? Y/N:\n")
    if loop == "Y":
        cipher.clear()
        output = ""
        program()

program()
```

```python
# Auction Program Flow Chart

# Take Name, Bid Value, and ask if there are other bidders
# If yes, clear screen

# It repeats the inputs
# It logs all the inputs

# Once no, it lists the winner and the bid value

# The clear function is native in replit, and I don't want to use a module so \n * 100

# Use the max() function, native, no math module needed

bidlist = []
bidvalues = []

def screen_clear():
    print("\n" * 100)

def new_bid():
    name = input("What is your name? \n")
    bidvalue = input("What is your bid? \n")
    loop = input("Are there other bidders? \n")

    bid_entry = {
        "name": name,
        "value": bidvalue,
    }

    bidlist.append(bid_entry)

    if loop == 'Y':
        screen_clear()
        new_bid()

print("Welcome to the Silent Auction")
new_bid()

bidlistlength = range(len(bidlist))

for x in bidlistlength:
    bidvalues.append(bidlist[x]["value"])

winvalue = max(bidvalues)

for x in bidlistlength:
    if bidlist[x]["value"] == winvalue:
        winnername = bidlist[x]["name"]

print(f"The winner is {winnername} with a bid of {winvalue}.")
print("This concludes the silent auction.")



    
```
